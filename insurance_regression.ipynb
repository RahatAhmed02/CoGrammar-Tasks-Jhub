Use the data in the relevant columns to determine how age affects insurance costs:
- Plot a scatter plot with age on the x-axis and charges on the y-axis.
- Using linear_model.LinearRegression() from sklearn, fit a model to your data, and make predictions on data.
- Plot another scatter plot with the best-fit line.

```python
# Import necessary libraries for data manipulation, visualisation, and linear regression modeling
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
```

```python
# Part a: Scatter plot with age on the x-axis and charges on the y-axis
# Load the insurance dataset from a CSV file
insurance = pd.read_csv("insurance.csv")

# Select only the 'age' and 'charges' columns for analysis
df = insurance[['age', 'charges']]

# Create a scatter plot to visualise the relationship between age and insurance charges
plt.figure()
sns.scatterplot(data=df, x='age', y='charges')
plt.show()
plt.close()

# Prepare the data for linear regression
x = df[['age']].values
y = df['charges'].values
```

```python
# Part b: Using linear_model.LinearRegression() from sklearn
# Create an instance of LinearRegression and fit to data
insurance_model = LinearRegression()
insurance_model.fit(x, y)

# Use the fitted model to make predictions on the original data
y_pred = insurance_model.predict(x)

# Part c: Scatter plot with the best-fit line
# Scatter plot of the actual data points
plt.scatter(x, y, color='b')
# Line plot of the model's predictions
plt.plot(x, y_pred, color='r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()
```

```python
# Making predictions for an unseen age
unk_x = [[90]]

# Append the new age to the existing age array
x_pred = np.append(x, unk_x).reshape(-1,1)
# Predict charges for all ages including the new unseen age
y_pred = insurance_model.predict(x_pred)

# Visualise the extended predictions including the new age
plt.scatter(x, y, color = 'b')
plt.plot(x_pred, y_pred, color = 'r')
plt.xlabel('Age')
plt.ylabel('Charges')
plt.show()

# Print the predicted charges for the new unseen age
print("Insurance charges should be:", insurance_model.predict(unk_x))
```
